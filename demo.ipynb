# Run prediction on one image

```python
%matplotlib inline
import matplotlib

import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt
from keras.applications.imagenet_utils import decode_predictions
from keras.preprocessing import image
from imageio import imread
from nasnet import mobile, preprocess

matplotlib.rcParams['figure.figsize'] = (10, 6)
```

### Load an image

```python
image = imread('images/peacock.jpg')
plt.imshow(image)
```

### Setup model, download weights and convert weights to Keras

```python
model = mobile(weights='imagenet')
```

### Preprocess the image

```python
size = model.input_shape[2]
x = np.expand_dims(preprocess(image, size=size), axis=0)
```

### Run the model

```python
y = model.predict(x)
```

### Top 5 results

```python
for index, res in enumerate(decode_predictions(y)[0]):
    print('{}. {}: {:.3f}%'.format(index + 1, res[1], 100 * res[2]))
```
